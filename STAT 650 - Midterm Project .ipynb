## 2. Data Analysis

### 2-1. Importing the libraries necessary for your EDA



```python
# Import your data easy to read
import numpy as np
import pandas as pd
```

### 2-2. Import libraries for EDA


```python
# Import libraries for EDA 
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2-3. Review Data


```python
data = pd.read_excel('nba_2022_2023.xlsx')
```

```python
# 2. Check types of data by .info() ; .dtype
data.info()
```

```python
# 3. For quantitative variable, generate a table for the count, mean, standard deviation, minimum and maximum values and the quantities of the data 
data.describe()
```

### 2-4. Generate Sub-dataset

```python
# 1. Make sub-dataset(s) from your original dataset for your research objective, goals by dropping unnecessary variables 

#Creating a sub data set where we have all the players that play the traditional 5 NBA Positions
desired_pos = ['C', 'SG', 'PF', 'PG', 'SF']

bol = []
for i in data['Pos']:
    if i in desired_pos:
        bol.append(True)
    else:
        bol.append(False)
```

```python
pos_data = data[bol].copy()
```

```python
ages = data.Age
```

```python
# 2. try to change variable names
new_column = []
for age in ages:
    if age>=19 and age<=25:
        new_column.append('19-25')
    elif age>=26 and age<=33:
        new_column.append('26-33')
    else:
        new_column.append('34-42')
```

```python
data['AgeGroups'] = new_column
```

```python
# 3. check if there are missing values
data.isnull().sum()
```

```python
#drop null values
data.dropna(inplace=True)
```

```python
# 4. Check if there are duplicated values
data.duplicated().sum()
```

### 2-5. Checking Outliers

```python
#gathering all the numeric columns for analysis
numeric_columns = ['G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
# check if quantitative variables have outliers.
fig, axes = plt.subplots(5, 5, figsize=(12, 12))

for i in range(5):
    for j in range(5):
        n = i*5 + j + 1
        axes[i, j].boxplot(data[numeric_columns[n-1]])
        axes[i, j].set_title(numeric_columns[n-1])
plt.tight_layout()
plt.show()
```

### 2-6. Generating Plot(s)

```python
# generate plots to support your objective and goals
#count plot for position
plt.rcParams["figure.figsize"] = (8, 4)
plt.rcParams["xtick.labelsize"] = 7
ax = sns.countplot(x = 'AgeGroups', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

You can see that there are more number of players between 19-25 and gradually it decreases and you can also observe that there are very few playing who are aged above 34 which seems to be true because having such a long career is rare.

```python
#lets check their performance
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
g = data.groupby('AgeGroups')
mean_data = g[numeric_columns].mean()
mean_data
```

lets plot the average performance of the players and see whether we can find some good insights.

```python
mean_data['AgeGroups'] = ['19-25','26-33','34-42']
```

```python
fig, axes = plt.subplots(5, 5, figsize=(12, 12))

for i in range(5):
    for j in range(5):
        n = i*5 + j + 1
        axes[i, j].bar(mean_data['AgeGroups'], mean_data[numeric_columns[n-1]])
        axes[i, j].set_title(numeric_columns[n-1])
plt.tight_layout()
plt.show()
```

you can see that aged people are good at skills such as accuracy of 2 Pointers, 3 Pointers. And we know that as age increased they tend to move very efficiently on court there are less number of 2P attempts and more number of 3P attempts. You can also see that the most number of middle aged people tend to start the games since they are well experienced and also they have good amount of stamina. Apart from these results there are many good trends that you can observe such as the trend in eFG%, ORB etc.

```python
#To check this lets start by analysing the number of players in each group
sns.countplot(x = 'Pos', data=pos_data)
plt.show()
```

you can observe that almost every position is being filled with equal number of players but the shooting guard position is being the position that seems to be popular

```python
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
g = pos_data.groupby('Pos')
mean_data = g[numeric_columns].mean()
mean_data
```

```python
mean_data['Pos'] = pos_data['Pos'].unique()
```

```python
fig, axes = plt.subplots(5, 5, figsize=(17, 17))

for i in range(5):
    for j in range(5):
        n = i*5 + j + 1
        axes[i, j].bar(mean_data['Pos'], mean_data[numeric_columns[n-1]])
        axes[i, j].set_title(numeric_columns[n-1])
plt.tight_layout()
plt.show()
```

You can see some interesting results. This plot basically tells us on what should a player be good at to be selected in that position so basically if one player wants to be a center he should be really good at Offensive and Defensive rebounds(ORB, DRB) and also should be good at scoring 2 pointers. Similarly if one player wants to be a Point Forward he should be good at his stamina because of MP column, assisting for baskets and also he should be good at stealing the ball from opponents. This is how you interpret the graph and check which player should be selected for a certain position

```python
#count plot for team
plt.rcParams["figure.figsize"] = (8, 4)
plt.rcParams["xtick.labelsize"] = 7
ax = sns.countplot(x = 'Tm', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
team_stats = data.groupby('Tm') [['PTS', 'AST', 'TRB']].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='PTS', y='Tm', data=team_stats, label='Points (PTS)', color='b')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='AST', y='Tm', data=team_stats, label='Assists (AST)', color='g')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TRB', y='Tm', data=team_stats, label='Rebounds (TRB)', color='r')
plt.show()
```

These graphs tells us on which teams perform the best accross all three aspects i.e Points, Assists and Total Rebounds
We can see that some of the teams like Boston Celtics performed well in all aspects. 

## 3. EDA


```python
# perform your EDA
#correlation heat map for all the numeric columns
correlation_matrix = data[numeric_columns].corr()
plt.figure(figsize=(20,16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

You can see that many of the columns are actually correlated to each other. Some of them are
- 3PA AND 3P
- 2PA AND 2P
- PTS AND AST etc

```python
#histogram for all the numeric columns
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(16, 12))
for i, column in enumerate(numeric_columns):
    row = i//5
    col = i%5
    sns.histplot(data[column],ax=axes[row,col], bins = 25, kde=True)
    axes[row, col].set_xlabel(column)
    
plt.tight_layout()
plt.show()
```

You can see that some of the data distributions actually follow really nice guassian distributions those are FG%, 2P%, 
eFG%

lets plot the scatter plots of the correlated attributes that we just mentioned above.
